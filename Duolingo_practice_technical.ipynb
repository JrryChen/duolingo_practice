```python
from typing import List

def phone_word_lookup(number: str) -> List[str]:
    '''
    use a recursive backtracking approach to find all possible letter combinations
    state: current string, current index
    action: add letter to current string, move to next index
    base case: if current index == len(number), add current string to result
    '''

    digit_to_letters = {
        '2': 'abc',
        '3': 'def',
        '4': 'ghi',
        '5': 'jkl',
        '6': 'mno',
        '7': 'pqrs',
        '8': 'tuv',
        '9': 'wxyz'
    }

    res = set()
    def backtrack(current: str, index: int):
        if index == len(number):
            res.add(current)
            return
        for letter in digit_to_letters.get(number[index], ''):
            backtrack(current + letter, index + 1)

    backtrack('', 0)
    return res        
```

```python
test_case1 = "2"
assert phone_word_lookup(test_case1) == set(['a', 'b', 'c'])

test_case2 = "1"
assert phone_word_lookup(test_case2) == set()

test_case3 = "23"
assert phone_word_lookup(test_case3) == set([
    'ad', 'ae', 'af',
    'bd', 'be', 'bf',
    'cd', 'ce', 'cf'
])

test_case4 = "19"
assert phone_word_lookup(test_case4) == set()

test_case5 = "81"
assert phone_word_lookup(test_case5) == set([])

print("All test cases pass")
```
